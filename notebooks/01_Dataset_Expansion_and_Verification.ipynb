# Notebook 1: Dataset Expansion and Verification

This notebook outlines the process for expanding the dataset with new gesture classes and verifying the integrity of the image data. This is the first step in retraining the model with more data.

### Process:
1.  **Plan New Classes**: Decide on the new gestures to add.
2.  **Update Label File**: Add the new class names to `keypoint_classifier_label.csv`.
3.  **Expand Image Dataset**: Create new folders and collect images for the new classes.
4.  **Verify Dataset**: Run a script to check for missing folders, incorrect image counts, and corrupted files.

### 1. Plan New Classes & Update Label File

First, decide on the new gestures you want to add (e.g., 'help', 'thank_you', 'stop').

Next, open the label file at `model_artifacts/keypoint_classifier_label.csv` and append the new classes, making sure to use contiguous IDs. For example:

```csv
0,afraid
1,agree
...
19,you
20,help
21,thank_you
```

### 2. Expand Image Dataset

For each new class, create a new folder in the `dataset` directory. Inside each new class folder, create subfolders for each user (`User_1`, `User_2`, etc.) and collect your images there. The structure should be:

```
dataset/
├── help/
│   ├── User_1/
│   └── User_2/
└── thank_you/
    ├── User_1/
    └── User_2/
```

You can use the image collection script from the previous `01_data_collection.ipynb` to capture new images.
