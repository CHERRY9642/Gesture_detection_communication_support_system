# Notebook 11: Final Model Selection
Goal: Select the best model for deployment


```python
import pandas as pd

results = {
    "Logistic Regression": 0.72,
    "SVM": 0.78,
    "Random Forest": 0.84,
    "Gradient Boosting": 0.86,
    "MLP": 0.91
}

df_results = pd.DataFrame.from_dict(
    results, orient="index", columns=["Accuracy"]
)

df_results
```

```python
df_results.plot(kind="bar", legend=False, figsize=(6,4))
```

```python
best_model = df_results["Accuracy"].idxmax()
best_model
```

```python
print(f"""
Final Decision:
- Selected Model: {best_model}
- Reason: Highest accuracy and stable performance
- Supports both single-hand and two-hand gestures
- Suitable for real-time deployment
""")
```
