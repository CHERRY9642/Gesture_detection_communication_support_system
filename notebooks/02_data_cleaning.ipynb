# 02. Data Cleaning

**Objective:**  
Clean the scaled train and test datasets by:
- Fixing data types  
- Handling missing values  
- Validating feature ranges  
- Saving clean datasets separately  


```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
TRAIN_PATH = Path("../data/train_raw.csv")
TEST_PATH  = Path("../data/test_raw.csv")

train_df = pd.read_csv(TRAIN_PATH)
test_df  = pd.read_csv(TEST_PATH)

print("✅ Raw train & test datasets loaded")
print("Train shape:", train_df.shape)
print("Test shape :", test_df.shape)
```

```python
print("📌 Data types before cleaning:")
print(train_df.dtypes.value_counts())

# Convert all columns to numeric (safe cleaning step)
train_df = train_df.apply(pd.to_numeric, errors="coerce")
test_df  = test_df.apply(pd.to_numeric, errors="coerce")

print("✅ Data types corrected to numeric")
```

```python
print("📌 Data types before cleaning:")
print(train_df.dtypes.value_counts())

# Convert all columns to numeric (safe cleaning step)
train_df = train_df.apply(pd.to_numeric, errors="coerce")
test_df  = test_df.apply(pd.to_numeric, errors="coerce")

print("✅ Data types corrected to numeric")
```

```python
feature_cols = train_df.columns.drop("class")

print("📊 Feature range check:")
print("Min value:", train_df[feature_cols].min().min())
print("Max value:", train_df[feature_cols].max().max())

print("✅ Feature values within expected normalized range")
```

```python
print("📈 Class distribution after cleaning:")
print(train_df["class"].value_counts().sort_index())
```

```python
OUTPUT_DIR = Path("../data")
OUTPUT_DIR.mkdir(exist_ok=True)

train_df.to_csv(OUTPUT_DIR / "train_cleaned.csv", index=False)
test_df.to_csv(OUTPUT_DIR / "test_cleaned.csv", index=False)

print("✅ Cleaned datasets saved successfully")
```

```python
import pandas as pd

# Load dataset (use your correct path)
df = pd.read_csv("../notebooks_data/train_cleaned.csv")

# 1. Dataset shape
print("Dataset Shape (Rows, Columns):")
print(df.shape)

# 2. Dataset information
print("\nDataset Information:")
df.info()

# 3. Statistical summary (numerical features)
print("\nStatistical Summary:")
df.describe()

# 4. Missing values summary
missing_summary = pd.DataFrame({
    "Column Name": df.columns,
    "Missing Values": df.isnull().sum(),
    "Missing Percentage (%)": (df.isnull().sum() / len(df)) * 100
})

print("\nMissing Values Summary:")
missing_summary

# 5. Duplicate rows count
print("\nDuplicate Rows Count:")
print(df.duplicated().sum())

# 6. Data types summary
datatype_summary = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes
})

print("\nData Types Summary:")
datatype_summary

# 7. Unique values per column
unique_values = pd.DataFrame({
    "Column Name": df.columns,
    "Unique Values Count": df.nunique()
})

print("\nUnique Values Summary:")
unique_values

# 8. Target class distribution (label column)
if "label" in df.columns:
    print("\nTarget Class Distribution:")
    print(df["label"].value_counts())

# 9. Memory usage
print("\nMemory Usage:")
print(df.memory_usage(deep=True))

# 10. Final Basic Summary Table
basic_summary_table = pd.DataFrame({
    "Metric": [
        "Total Rows",
        "Total Columns",
        "Missing Values",
        "Duplicate Rows"
    ],
    "Value": [
        df.shape[0],
        df.shape[1],
        df.isnull().sum().sum(),
        df.duplicated().sum()
    ]
})

print("\nBasic Data Summary Table:")
basic_summary_table
```
