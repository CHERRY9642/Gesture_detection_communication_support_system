```python
# Load required libraries
import pandas as pd
import numpy as np
from pathlib import Path

print("📂 Loading keypoint dataset...")

# Load dataset
df = pd.read_csv("../model_artifacts/keypoint.csv")

# Show basic info
print("✅ Dataset loaded successfully")
print(f"🔢 Shape (rows, columns): {df.shape}")

df.head()
```

```python
print("🔍 Checking for missing values...")

missing_count = df.isnull().sum().sum()

print(f"❓ Total missing values in dataset: {missing_count}")
```

```python
print("🛠️ Cleaning missing values and fixing data types...")

# Replace NaN with 0
df.fillna(0, inplace=True)

# Ensure correct data types
df["class"] = df["class"].astype(int)
df.iloc[:, 1:] = df.iloc[:, 1:].astype(np.float32)

print("✅ Missing values handled")
print("✅ Data types corrected")

print(df.dtypes.head())
```

```python
print("🧹 Removing invalid samples (no hand landmarks)...")

# Extract feature matrix
features = df.iloc[:, 1:].values

# Identify rows where BOTH hands are zero vectors
mask = [
    not (np.all(row[:42] == 0) and np.all(row[42:] == 0))
    for row in features
]

before = df.shape[0]
df = df[mask]
after = df.shape[0]

print(f"❌ Invalid samples removed: {before - after}")
print(f"✅ Remaining valid samples: {after}")
```

```python
print("💾 Saving cleaned dataset...")

# Create directory if not exists
Path("../data/processed").mkdir(parents=True, exist_ok=True)

# Save cleaned data
output_path = "../data/processed/clean_keypoint.csv"
df.to_csv(output_path, index=False)

print(f"✅ Clean dataset saved at: {output_path}")
print(f"📊 Final dataset shape: {df.shape}")
```
