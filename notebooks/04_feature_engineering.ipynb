# Notebook 04: Feature Engineering
Goal: Improve landmark representation for ML models


```python
import pandas as pd
import numpy as np

print("📂 Loading cleaned keypoint dataset...")

df = pd.read_csv("../data/processed/clean_keypoint.csv")

print("✅ Dataset loaded successfully")
print(f"🔢 Shape (rows, columns): {df.shape}")

df.head()
```

```python
print("✋ Separating right-hand and left-hand features...")

# First 42 → Right hand, Next 42 → Left hand
right_hand = df.iloc[:, 1:43].values
left_hand = df.iloc[:, 43:85].values

print(f"👉 Right hand shape: {right_hand.shape}")
print(f"👉 Left hand shape : {left_hand.shape}")
```

```python
print("🧠 Computing hand presence features...")

# 1 if hand is present, else 0
right_present = (np.abs(right_hand).sum(axis=1) > 0).astype(int)
left_present  = (np.abs(left_hand).sum(axis=1) > 0).astype(int)

print("✅ Hand presence calculated")
print("Sample (Right, Left):")
print(list(zip(right_present[:10], left_present[:10])))
```

```python
print("⚡ Calculating hand energy features...")

# Energy = magnitude of landmark movement
right_energy = np.linalg.norm(right_hand, axis=1)
left_energy  = np.linalg.norm(left_hand, axis=1)

print("✅ Energy features computed")
print("Sample energies (Right, Left):")
print(list(zip(right_energy[:5], left_energy[:5])))
```

```python
print("🧩 Creating feature-engineered dataset (analysis only)...")

df_fe = df.copy()

df_fe["right_hand_present"] = right_present
df_fe["left_hand_present"] = left_present
df_fe["right_hand_energy"] = right_energy
df_fe["left_hand_energy"] = left_energy

print("✅ Feature engineering completed")
print(f"🔢 New shape: {df_fe.shape}")

df_fe.head()
```

```python
from pathlib import Path

print("💾 Saving feature-engineered dataset...")

Path("../data/processed").mkdir(parents=True, exist_ok=True)

output_path = "../data/processed/feature_engineered_keypoint.csv"
df_fe.to_csv(output_path, index=False)

print(f"✅ Saved at: {output_path}")
print(f"📊 Final dataset shape: {df_fe.shape}")
```
