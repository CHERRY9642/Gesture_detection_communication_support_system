# 04. Feature Engineering

**Objective:**  
Create additional meaningful features from hand landmark data to improve
model performance and robustness.


```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
TRAIN_PATH = Path("../data/train_cleaned.csv")
TEST_PATH  = Path("../data/test_cleaned.csv")

train_df = pd.read_csv(TRAIN_PATH)
test_df  = pd.read_csv(TEST_PATH)

print("✅ Clean datasets loaded")
print("Train shape:", train_df.shape)
print("Test shape :", test_df.shape)
```

```python
X_train = train_df.drop(columns=["class"])
y_train = train_df["class"]

X_test = test_df.drop(columns=["class"])
y_test = test_df["class"]

print("Features shape:", X_train.shape)
print("Labels shape  :", y_train.shape)
```

```python
def add_engineered_features(X):
    X_new = X.copy()

    # 1️⃣ Mean of all landmark coordinates
    X_new["feat_mean"] = X.mean(axis=1)

    # 2️⃣ Standard deviation of landmarks
    X_new["feat_std"] = X.std(axis=1)

    # 3️⃣ Max landmark distance (spread of hand)
    X_new["feat_max"] = X.max(axis=1)

    # 4️⃣ Min landmark distance
    X_new["feat_min"] = X.min(axis=1)

    # 5️⃣ Range (max - min)
    X_new["feat_range"] = X_new["feat_max"] - X_new["feat_min"]

    return X_new
```

```python
X_train_fe = add_engineered_features(X_train)
X_test_fe  = add_engineered_features(X_test)

print("✅ Feature engineering applied")
print("New feature count:", X_train_fe.shape[1] - X_train.shape[1])
```

```python
print("📊 Engineered feature summary:")
X_train_fe.iloc[:, -5:].describe().T
```

```python
OUTPUT_DIR = Path("../data")
OUTPUT_DIR.mkdir(exist_ok=True)

train_fe = X_train_fe.copy()
train_fe["class"] = y_train.values
train_fe.to_csv(OUTPUT_DIR / "train_fe.csv", index=False)

test_fe = X_test_fe.copy()
test_fe["class"] = y_test.values
test_fe.to_csv(OUTPUT_DIR / "test_fe.csv", index=False)

print("✅ Feature-engineered datasets saved")
```
